```python
import pandas as pd  
import numpy as  np 
import matplotlib.pyplot as plt
```

# 1. Cleaning the Data 

```python
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]
sales_19df = pd.DataFrame()
for month in months:
    temp_df = (pd.read_csv(r"C:\Users\Cliff Chen\Documents\Python\Sales Data Project\Pandas-Data-Science-Tasks-master" + 
                            "\SalesAnalysis\Sales_Data\Sales_" + month + "_2019.csv"))
    sales_19df = sales_19df.append(temp_df, ignore_index=True)

sales_19df.columns
```

```python
sales19_ord = sales_19df.sort_values(by=["Order Date"])
sales19_ord["Order Date"].str.contains("/20 ")
    
sales19_ord
```

```python
sales19_ord = sales19_ord.dropna()

sales19_ord
```

```python
sales_data = sales19_ord[sales19_ord.Product != "Product"]
sales_data
```

```python
sales_data = sales_data.astype({"Price Each": float})

sales_data
```

```python
sales_data = sales_data.astype({"Quantity Ordered": int})
```

# 2. Add Month Column 

```python
sales_data["Month"] = sales_data["Order Date"].str[0:2]
sales_data = sales_data.astype({"Month": int})

sales_data.head()
```

# 3. Add Sales Column

```python
sales_data["Sales"] = sales_data["Quantity Ordered"] * sales_data["Price Each"]

sales_data.head()
```

```python
sales_data.columns = [title.lower() for title in sales_data.columns]

sales_data.head()
```

### Question 1: What was the best month for sales? How much was earned that month?

```python
monthly_sales = sales_data.groupby("month").sum()
# monthly_sales = monthly_sales.sort_values("sales", ascending=False) # - sorts by highest sales month

monthly_sales 
```

```python
monthly_sales["avg price per item"] = monthly_sales["sales"] / monthly_sales["quantity ordered"]

monthly_sales
```

```python
month_num = range(1,13)

plt.figure(figsize=(14,8))
plt.bar(month_num, monthly_sales["sales"])
plt.xticks(month_num)
plt.ylabel("Sales ($)") 
plt.xlabel("Month (Digit)")
plt.title("Total Sales Per Month (2019)")
plt.show()

# december sells the most, presumably for christmas | january sells the least, presumably after christmas
```

### Question 2: What city had the highest number of sales? 

```python
def get_city_state(address): 
    city = address.split(",")[1]
    state = address.split(",")[2].split(" ")[1]
    cityState = city + ", " + state
    return cityState

sales_data["location"] = sales_data["purchase address"].apply(lambda x: get_city_state(x))
#sales_data = sales_data.drop(columns="location")

sales_data.head()
```

```python
city_sales = sales_data.drop(columns=["month","price each"]).groupby("location").sum()

city_sales
```

```python
#cities = sales_data["location"].unique() # stores data in diff order than data
cities = [city for city, df in sales_data.groupby("location")]

plt.figure(figsize=(16,8))
plt.bar(cities, city_sales["sales"])
plt.xticks(cities,rotation="vertical", size=10)
plt.ylabel("Sales ($)") 
plt.xlabel("Location (City, State)")
plt.title("Total Sales Per City (2019)")

plt.show()

# portland sells the least, more advertising | san fran sells the most, more advertising | what to do with in between? 
```

### Question 3: What time to Display Advertisements? 


```python
sales_data["order date"] = pd.to_datetime(sales_data["order date"])
```

```python
sales_data["order time"] = round(sales_data["order date"].dt.hour + sales_data["order date"].dt.minute/60, 3)

sales_data
```

```python
sales_data["count"] = 1
sales_data["hour"] = sales_data["order date"].dt.hour

hour = [hour for hour, df in sales_data.groupby("hour")]
sales_data = sales_data.drop(columns="hour")


time = [time for time, df in sales_data.groupby("order time")]
time_sales = sales_data.groupby("order time").count()

time_sales
```

```python
plt.figure(figsize=(20,10))
plt.plot(time, time_sales["count"])
plt.xticks(hour, size=10)
plt.ylabel("Number of Sales") 
plt.xlabel("Time (Hour)")
plt.title("Total Sales Over the Day (2019)")
plt.grid()

plt.show()

# when to run ads to sell more items
```

### Question 4: What Products are Most Often Sold Together? 

```python
group_sales = sales_data[sales_data["order id"].duplicated(keep=False)] # filter for duplicated order id (multiple purchases)

group_sales
```

```python
group_sales["group"] = group_sales.groupby("order id")["product"].transform(lambda x: ", ".join(x))

group_sales
```

```python
group_sales = group_sales[["order id", "group"]].drop_duplicates()

group_sales
```

```python
group_sales["group"]
```

```python
from itertools import combinations 
from collections import Counter 

count = Counter()

for row in group_sales["group"]: 
    grp_list = sorted(row.split(", "), key=str.lower)
    count.update(Counter(combinations(grp_list, 2)))
    
   
count.most_common(10)

# can decide which items to put together in a store
```
